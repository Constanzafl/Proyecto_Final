```python
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lex_rank import LexRankSummarizer
```

```python
import nltk
nltk.download('punkt')
```

aplico sumy para resumir el texto de la reseña

```python
# Crear una función para resumir una reseña
def resumir_reseña(reseña):
    parser = PlaintextParser.from_string(reseña, Tokenizer("english"))
    summarizer = LexRankSummarizer()
    summary = summarizer(parser.document, 2)  # Cambia el valor a la cantidad de oraciones deseadas en el resumen
    return ' '.join([str(sentence) for sentence in summary])

# Aplicar la función de resumen a la columna 'reseñas' y almacenar los resúmenes en una nueva columna
resumen_dfcompleto['resumen_reseñas'] = resumen_dfcompleto['text'].apply(resumir_reseña)
```

```python
latitud_usuario = 26.710094
longitud_usuario = -80.106882
# Definir el radio de 2 km
radio_km = 3

# Función para filtrar lugares dentro del radio especificado
def filtrar_lugares_cercanos(resumen_dfcompleto, lat_user, lon_user, radio_km):
    def calcular_distancia(row):
        lugar_lat = row['latitude_x']
        lugar_lon = row['longitude_x']
        distancia = geodesic((lat_user, lon_user), (lugar_lat, lugar_lon)).kilometers
        return distancia

    resumen_dfcompleto['distancia'] = resumen_dfcompleto.apply(calcular_distancia, axis=1)
    lugares_cercanos2 = resumen_dfcompleto[resumen_dfcompleto['distancia'] <= radio_km]

    return lugares_cercanos2

# Filtrar los lugares dentro del radio especificado
lugares_cercanos2 = filtrar_lugares_cercanos(resumen_dfcompleto, latitud_usuario, longitud_usuario, radio_km)

# Mostrar los lugares cercanos
lugares_cercanos2.head(16)
```
