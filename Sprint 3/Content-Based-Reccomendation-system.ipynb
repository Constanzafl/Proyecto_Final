```python
df_4 = filtered_df

# Joining multiple columns into a new column
df_4['All_Cols'] = filtered_df[['category', 'Offerings', 'Atmosphere', 'Highlights']].apply(lambda row: ', '.join(map(str, row)), axis=1)


df_4.head()
```

```python
df_4['All_Cols'] = df_4['All_Cols'].map(lambda x: str(x))
df_4['All_Cols'] = df_4['All_Cols'].map(lambda x: x.lower())
print(df_4.shape)
df_4.head()
```

```python
df4_final = df_4.groupby('name_y')['All_Cols'].sum()
df4_final = df4_final.to_frame(name = 'sum').reset_index()
df4_final
```

```python
# Getting a list of Unique Keywords per Restaurant

df4_final['sum'] = df4_final['sum'].map(lambda x: x.replace(", ","', '"))
df4_final['sum'] = df4_final['sum'].map(lambda x: str("'") + x + str("'"))
f = lambda x: x["sum"].split(", ")
df4_final['sum'] = df4_final.apply(f, axis=1)
df4_final['sum'] = df4_final['sum'].map(lambda x: set(x))
df4_final.set_index('name_y', inplace = True)
df4_final.head()
```

```python
# Remove quotation marks
df4_final['sum'] = df4_final['sum'].astype(str)  # Ensure the column contains strings

df4_final['sum'] = df4_final['sum'].map(lambda x: x.replace("'", ""))
df4_final['sum'] = df4_final['sum'].map(lambda x: x.replace("{", ""))
df4_final['sum'] = df4_final['sum'].map(lambda x: x.replace("}", ""))
df4_final.head()
```

```python
df4_final.head()
```

```python
count = CountVectorizer()
count_matrix = count.fit_transform(df4_final['sum'])

indices = pd.Series(df4_final.index)
indices[:5]
```

```python
count_matrix.shape
```

```python
cosine_sim = cosine_similarity(count_matrix, count_matrix)
cosine_sim
```

```python
cosine_sim.shape
```

```python

# function that takes in movie title as input and returns the top 10 recommended movies
def content_based_recommendations(name, cosine_sim = cosine_sim):
    
    recommended_restaurants = []
    
    # gettin the index of the movie that matches the title
    idx = indices[indices == name].index[0]

    # creating a Series with the similarity scores in descending order
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)

    # getting the indexes of the 10 most similar movies
    top_10_indexes = list(score_series.iloc[1:11].index)
    
    # populating the list with the titles of the best 10 matching movies
    for i in top_10_indexes:
        recommended_restaurants.append(list(df4_final.index)[i])
        
    return recommended_restaurants
```

```python
content_based_recommendations('10x vegan')
```
