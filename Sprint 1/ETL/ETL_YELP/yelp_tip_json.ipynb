```python
import pandas as pd

chunk_iter = pd.read_json("datasets/tip.json", lines=True, chunksize=1000)

# Usamos una lista para guardar cada chunk
list_of_chunks = []

# Iteramos sobre cada chunk y lo agregamos a la lista
for chunk in chunk_iter:
    list_of_chunks.append(chunk)

# Concatenamos todos los chunks en un único DataFrame
df = pd.concat(list_of_chunks, axis=0)
```

```python
# Ajustar opciones de visualización
# Mostrar solo los primeros 50 caracteres de cada campo
pd.set_option('display.max_colwidth', 50)
pd.set_option('display.max_rows', False)
pd.set_option('display.max_columns', False)
```

```python

# 1. Lee el archivo CSV en un DataFrame
df_business = pd.read_csv("business.csv")

# 2. Realiza el join "left"
merged_df = df_business.merge(df, on="business_id", how="left", suffixes=('', '_right'))

# 3. Elimina la columna "business_id_right" si existe
if "business_id_right" in merged_df.columns:
    merged_df.drop("business_id_right", axis=1, inplace=True)

# 4. Identifica cuántas filas tienen valores faltantes en las columnas del DataFrame de la derecha
# (Considerando que las columnas del DataFrame de la derecha en el merged_df son 'user_id', 'text', 'date', 'compliment_count')
missing_values = merged_df['user_id'].isnull().sum()

print(f"Hay {missing_values} 'business_id' en el archivo business.csv que no tienen coincidencia en el DataFrame 'df'.")
```

```python

# Función para eliminar los strings "\n"
def replace_newlines(value):
    if isinstance(value, dict):
        return {k: v.replace("\n", "") for k, v in value.items()}
    elif isinstance(value, str):
        return value.replace("\n", "")
    return value

# Aplicar la función a cada celda del DataFrame
merged_df = merged_df.applymap(replace_newlines)
```
