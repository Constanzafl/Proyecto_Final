```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')


# Ajustar opciones de visualización
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Cargar el DataFrame desde un archivo .pkl
df = pd.read_json(r'datasets/yelp_academic_dataset_business.json', lines=True)


df.head(1)
```

```python
df.shape
```

```python
df.info()
```

```python
df = df.applymap(lambda x: str(x) if isinstance(x, dict) else x)
df = df.drop_duplicates()
```

```python
df.info()
```

```python
# Dividimos el DataFrame en dos: uno con las columnas originales y otro con las duplicadas
df_original = df.iloc[:, :14]
df_duplicado = df.iloc[:, 14:]

# Reemplazamos los valores faltantes en df_original con los valores de df_duplicado
for col in df_original.columns:
    df_original[col].fillna(df_duplicado[col], inplace=True)

# Eliminamos las columnas duplicadas
df = df_original
```

```python
df.info()
```

```python
"""# Condición 1: state es 'FL'
mask_state = df['state'] == 'FL'

# Condición 2: la columna 'categories' contiene el string 'Restaurants'
# Usamos el método str.contains() y aseguramos que no haya NaNs usando fillna("")
mask_restaurants = df['categories'].fillna("").str.contains('Restaurants')

# Condición 3: la columna 'city' es 'Miami'
#mask_city = df['city'] == 'Miami'

# Combinamos todas las condiciones con el operador "and" (&)
combined_mask = mask_state & mask_restaurants #& mask_city

# Filtramos el DataFrame
filtered_df = df[combined_mask]
"""
# Importando la librería necesaria
from uszipcode import SearchEngine

# Inicializa el buscador
search = SearchEngine()

# Obtiene una lista de todos los códigos postales de Florida
zipcodes_florida = search.by_state("Florida", returns=0)

# Extrae las ciudades únicas
cities_florida = set([zipcode.major_city for zipcode in zipcodes_florida])
cities_florida_lower = [city.lower() for city in cities_florida]

# Filtra el DataFrame usando la lista de ciudades
mask = df['city'].str.lower().isin(cities_florida_lower)
filtered_df = df[mask]
```
